```python
import pandas as pd
import json

# Load your CSV
df = pd.read_csv("data/akc-data-2020-05-18 11_36_36.944067.csv")

# Define mappings (example: map categories to 1–3 scale)
energy_map = {"low": 1, "medium": 2, "high": 3}
grooming_map = {"low": 1, "medium": 2, "high": 3}
trainability_map = {"low": 1, "medium": 2, "high": 3}
demeanor_map = {"reserved": 1, "balanced": 2, "outgoing": 3}  # Adjust to your actual data
```

```python
df
```

```python
# Helper to safely map with a fallback
def safe_map(value, mapping):
    return mapping.get(str(value).lower().strip(), 2)
```

```python
# Create a simplified traits dict for each breed
breed_list = []
for _, row in df.iterrows():
    breed = {
        "name": row["breed"] if "breed" in row else row["name"],
        "traits": {
            "energy": safe_map(row["energy_level_category"], energy_map),
            "grooming": safe_map(row["grooming_frequency_category"], grooming_map),
            "trainability": safe_map(row["trainability_category"], trainability_map),
            "demeanor": safe_map(row["demeanor_category"], demeanor_map),
            # Simple size logic using weight
            "size": 3 if row["max_weight"] > 60 else 2 if row["max_weight"] > 30 else 1
        }
    }
    breed_list.append(breed)
```

```python
# Save to JSON
with open("breeds.json", "w") as f:
    json.dump(breed_list, f, indent=2)
```
